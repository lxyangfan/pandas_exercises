### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Create 3 differents Series, each of length 100, as follows: 
1. The first a random number from 1 to 4 
2. The second a random number from 1 to 3
3. The third a random number from 10,000 to 30,000

```python
first = pd.Series( np.random.randint(1, high=5,size=100 ) )
sec = pd.Series(np.random.randint(1, high=4, size=100 ) )
third = pd.Series( np.random.randint(10000, high=30001, size=100 ))
```

### Step 3. Let's create a DataFrame by joinning the Series by column

```python
data = pd.concat([first, sec, third], axis=1)
data
```

### Step 4. Change the name of the columns to bedrs, bathrs, price_sqr_meter

```python
data.rename(columns={0: 'bedrs', 1:'abthrs', 2: 'price_sqr_meter'}, inplace= True)
data.head()
```
